```python
# Dictionary to standardize Configuration names
config_map = {
    '3Bhk': '3BHK',
    '3bhk': '3BHK',
    '3BHK': '3BHK',

    '4bhk': '4BHK',
    '4Bhk': '4BHK',
    '4BHK': '4BHK',

    '5bhk+': '5BHK+',
    '5Bhk+': '5BHK+',
    '5BHK+': '5BHK+',
}

# Apply the mapping to standardize Configuration names
clean_df['Configuration'] = clean_df['Configuration'].map(config_map)
```

```python
# 1. Price per Sqft
clean_df['Price_per_Sqft'] = (clean_df['Ticket_Price_Cr'].abs() * 10000000)  / clean_df['Unit_Size_Sqft']  # 1 Cr = 1e7

# 2. Quarter Number (from Purchase_Quarter)
clean_df['Quarter_Number'] = clean_df['Purchase_Quarter'].dt.quarter

```
